### Content based filtering (Mathematical approach, no model)

By calculating the average genre profile for a user based on the movies they rated

```python
'''
def get_film_genres(genres_df, film_idx) -> list:
    
    genres = genres_df.iloc[film_idx, 2:]

    genres = genres[genres == 1]

    return genres.index.tolist()
'''


def user_profile(reviews_df, genres_df, user_idx) -> pd.Series:

    #name = reviews_df.iloc[user_idx, 1]

    reviews = reviews_df.iloc[user_idx, 2:]

    rated_movies = reviews[reviews != 0]
    unrated_movies = reviews[reviews == 0]

    genres_rated_movies = genres_df.loc[rated_movies.index]
    genres_unrated_movies = genres_df.loc[unrated_movies.index]

    
    profile = genres_rated_movies.T.dot(rated_movies)
    profile /= profile.sum() #normalize
    print(profile)

    similarity_score = genres_unrated_movies.dot(profile)
    return similarity_score
```

```python
#Vincent
user_profile(user_reviews_copy, movie_genres_copy, 0)
```

```python
users = ['Vincent', 'Edgar', 'Addilyn', 'Marlee', 'Javier']

for i, user in enumerate(users):

    user_scores = user_profile(user_reviews_copy, movie_genres_copy, i)

    recommended_movies = user_scores.sort_values(ascending=False).head(5)

    recommended_movies = recommended_movies.index.tolist()

    print(f'Movie recommendations for {user}:\n {recommended_movies}\n')
```
